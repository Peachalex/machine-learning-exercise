```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
boston_data =  pd.read_csv("BostonHousing.csv")
boston_data.head()
```

```python
boston_data.describe()
```

```python
y = boston_data.medv
```

```python
X = boston_data.drop(["medv"],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
X_train.shape, y_test.shape
```

```python
X_train.shape, y_train.shape
```

```python
boston_model = linear_model.LinearRegression()
```

```python
boston_model.fit(X_train, y_train)
```

```python
y_predict = boston_model.predict(X_test)
```

```python
print("Coefficients:", boston_model.coef_)
print("Intercept: ", boston_model.intercept_)
print("Mean Squared Error : %.2f " % mean_squared_error(y_test, y_predict))
print("Coefficient of determination (R Square score): %.2F" % r2_score(y_test, y_predict))
```
