```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
```

```python
#create normal distribution dataset with mean 0 and standard deviation 1
#plot it
data = np.random.normal(0,1,100)
data_normal = pd.Series(data)
fig, ax = plt.subplots()
data_normal.plot.hist(ax=ax, bins=25, density=True)
data_normal.plot.kde(ax=ax)
plt.show()
```

```python
#QQ plot of normal distribution vs theoritical normal ditribution
fig = sm.qqplot(data_normal, line='45',fit=True, dist=stats.norm)
plt.show()
```

```python
#creating uniform dataset from -2 to 2 and plot it
data_uniform = pd.Series(np.random.uniform(-2,2,100))
fig, ax = plt.subplots()
data_uniform.plot.hist(ax=ax, bins=25, density=True)
data_uniform.plot.kde(ax=ax)
plt.show()
```

```python
#QQ plot of uniform distribution against with normal distribution
fig_uni = sm.qqplot(data_uniform,line='45', fit=True, dist=stats.norm)
plt.show()
```

```python
#create exponential dataset and plot it
data_exp = pd.Series(np.random.exponential(scale=1.0, size=100))
fig, ax = plt.subplots()
data_exp.plot.hist(ax=ax, bins=25, density=True)
data_exp.plot.kde(ax=ax)
plt.show()
```

```python
#qqplot for exponential distribution with exponential distribution
data_exp = sm.qqplot(data_exp, line='45', fit=True, dist=stats.expon)
plt.show()
```

```python
#create left skewed distribution dataset
from scipy.stats import skewnorm
skewness = -5 # negaive number means left skewed 
data_random = skewnorm.rvs(a=skewness, loc=1000, size=50000)
data_random = data_random-min(data_random)
data_random = data_random / max(data_random)
data_random = data_random * 50000
data_random = pd.Series(data_random)
```

```python
#plot left skewed plot
fig, ax = plt.subplots()
data_random.plot.hist(ax=ax, bins=25, density=True)
data_random.plot.kde(ax=ax)
plt.show()
```

```python
fig_leftskewed = sm.qqplot(data_random, line='45', fit=True,dist=stats.norm)
plt.show()
```

```python
#create right skewed dataset
skewness = 12 # postive number means right skewed
data_random = skewnorm.rvs(a=skewness, loc=1000, size=50000)
data_random = data_random-min(data_random)
data_random = data_random / max(data_random)
data_random = data_random * 50000
data_random = pd.Series(data_random)
```

```python
#plot right skewd distribution
fig, ax = plt.subplots()
data_random.plot.hist(ax=ax, bins=25, density=True)
data_random.plot.kde(ax=ax)
plt.show()
```

```python
#qq plot fro right skewed distribution with normal distribution
fig_rightskewed = sm.qqplot(data_random, line='45', fit=True, dist=stats.norm)
plt.show()
```
